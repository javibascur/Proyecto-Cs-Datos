## 2023

```python
#Leemos el csv y pasamos a un dataframe
df_2023 = pd.read_csv("datos_utilizados/rendicion/2023/ArchivoC_Adm2023.csv", sep=";", low_memory=False)
```

```python
#Eliminamos las columnas que no utilizaremos
df_2023 = df_2023.drop(columns=["MATE1_REG_ANTERIOR", "HCSOC_REG_ANTERIOR", "CIEN_REG_ANTERIOR", "MODULO_REG_ANTERIOR", "CLEC_REG_ANTERIOR", "RBD", "COD_ENS", "MODULO_INV_ACTUAL", "CIEN_INV_ACTUAL", "HCSOC_INV_ACTUAL", "MATE_INV_ACTUAL", "CLEC_INV_ACTUAL", "MODULO_REG_ACTUAL", "PORC_SUP_NOTAS", "RAMA_EDUCACIONAL", "SITUACION_EGRESO"])
df_2023.head()
```

```python
#Revisamos si lso datos están en el Dtype que queremos
df_2023.info()
```

```python
#Vemos como str los datos de cada columna para reemplazar "," por "." y poder pasarlos a numeros "float"
df_2023["PROMEDIO_NOTAS"] = df_2023["PROMEDIO_NOTAS"].str.replace(',', '.').astype(float)

#Agrego la columna del año correspondiente de los datos para diferenciarlos en el dataframe final
df_2023["AÑO"] = 2023

#Así tenemos el id con el numero que se necesita
df_2023["ID_aux"] = df_2023["ID_aux"].str[3:].astype(str)
```

```python
#Reorganizamos el dataframe para que el año quede al inicio
df_2023 = df_2023[["AÑO","ID_aux","GRUPO_DEPENDENCIA","CODIGO_REGION","CODIGO_COMUNA","PROMEDIO_NOTAS","PTJE_NEM","PTJE_RANKING","CLEC_REG_ACTUAL","MATE1_REG_ACTUAL","MATE2_REG_ACTUAL","HCSOC_REG_ACTUAL","CIEN_REG_ACTUAL"]]
df_2023.head()
```

```python
#vuelvo a revisar que todos los cambios se hayan completado satisfactoriamente
df_2023.info()
```

**AHORA UNIMOS TODOS LOS DATOS EN UN MISMO DATAFRAME**

```python
df_final = pd.concat([df_2023, df_2022, df_2021, df_2020, df_2019], axis=0, join="outer").reset_index(drop=True)
```

```python
#Y eliminamos todos los datos que podrian afectarnos en los codigos futuros
df_final = df_final.replace('*', np.nan)
df_final = df_final.replace('None', np.nan)
df_final = df_final.replace(' ', np.nan)
df_final = df_final.replace(-999, np.nan)
df_final = df_final.replace('', np.nan)
df_final
```
