```python
import pandas as pd
```

## 2023

```python
df_2023 = pd.read_csv("datos_utilizados/matricula/2023/ArchivoMatr_Adm2023.csv", sep=";", low_memory=False)
```

```python
conservar = ["ID_aux","VIA","PTJE_POND"]
df_2023 = df_2023.drop(columns=[col for col in df_2023.columns if col not in conservar])
```

```python
df_2023
```

```python
# Agregamos la columna de año 
df_2023['AÑO'] = 2023
df_2023.head()
```

```python
# Editamos el 'ID_aux' de los alumnos para que solo quede el digito
df_2023['ID_aux'] = df_2023['ID_aux'].str[3:].astype(str)

print(df_2023['ID_aux'].dtype)
print(df_2023['VIA'].dtype)
print(df_2023['PTJE_POND'].dtype)
print(df_2023['AÑO'].dtype)
```

```python
df_2023.head()
```
