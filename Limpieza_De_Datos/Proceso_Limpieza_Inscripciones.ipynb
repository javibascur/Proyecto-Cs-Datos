```python
# De entre las columnas que quedaron afortunadamente los tipos de los datos estan colocados de la forma correcta, quedando como:
```

```python
### ID_aux: codigo unico que identifica a cada usuario (object)
### ANYO_PROCESO: a√±o en que se efectua la prueba a la que se inscribio el alumno (int)
### ANYO_EGRESO: A√±o en que el postulante egres√≥ de la educaci√≥n media (int)
### GRUPO_DEPENDENCIA: Tipo de financiamiento del establecimiento de egreso (float)
### CODIGO_REGIN_D: C√≥digo de la regi√≥n de domicilio del postulante (int)
### CODIGO_COMUNA_D: C√≥digo de la comuna de domicilio del postulante (int)
### SEXO: Sexo del postulante, esta planteado como categorias (1 para masculino y 2 para femenino) (int)
### FECHA_NACIMIENTO: Mes y a√±o de nacimiento del postulante (int)
### CON_QUIEN_VIVE: cantidad de personas con las que vive la persona (int)
### INGRESO_BRUTO_FAM: ingreso bruto mensual del grupo familiar del estudiante (int)
### INGRESO_MENSUAL_TOTAL: ingreso mensual del grupo familiar del postulante (float)L
### INGRESO_PERCAPITA_GRUPO_FA: Ingreso per c√°pita del grupo familiar del postulant (float)
```

```python
# Ahora lo que se quiere es pasar convertir las columnas de CON_QUIEN_VIVE e INGRESO_MENSUAL_TOTAL o INGRESO_BRUTO_FAM (segun corresponda)
# a INGRESO_PERCAPITA_GRUPO_FA para que asi todos los a√±os tengan las mismas columnas

# La manera en que esto se calculara sera dividiendo el ingreso familiar (segun corresponda) entre la cantidad de personas del grupo
```

```python
df2020["INGRESO_PERCAPITA_GRUPO_FA"] = df2020["INGRESO_MENSUAL_TOTAL"]//df2020["CON_QUIEN_VIVE"]
df2020 = df2020.drop(columns=[("INGRESO_MENSUAL_TOTAL"), ("CON_QUIEN_VIVE")])
```

```python
df2020.info()
```

```python
df2019["INGRESO_PERCAPITA_GRUPO_FA"] = df2019["INGRESO_BRUTO_FAM"]//df2019["CON_QUIEN_VIVE"]
df2019 = df2019.drop(columns=[("INGRESO_BRUTO_FAM"), ("CON_QUIEN_VIVE")])
```

```python
df2019.info()
```

```python
# Se unen los df de todos los a√±os con el metodo .concat para manejar mejor los datos
```

```python
dfUnido = pd.concat([df2023, df2022, df2021, df2020, df2019], axis=0, join="outer").reset_index(drop=True)
dfUnido
```

```python
# Reemplazamos los posibles valores "vacios con NaN pero no los eliminaremos, ya que se eliminaria gran cantidad de datos 
# importantes y  nuestro interes es saber cuales son nulos
dfUnido = dfUnido.replace('*', np.nan)
dfUnido = dfUnido.replace('None', np.nan)
dfUnido = dfUnido.replace(' ', np.nan)
dfUnido = dfUnido.replace(-999, np.nan)
```

# Filtraciones y Cambios Generales

```python
# Cambiamos algunos tipos de datos
dfUnido['GRUPO_DEPENDENCIA'] = dfUnido['GRUPO_DEPENDENCIA'].fillna(0).astype(int).astype('category')
dfUnido['INGRESO_PERCAPITA_GRUPO_FA'] = dfUnido['INGRESO_PERCAPITA_GRUPO_FA'].fillna(0).astype(int).astype('category')
dfUnido
```
