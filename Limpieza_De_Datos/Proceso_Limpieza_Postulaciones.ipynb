```python
# se cambia la primera parte de los codigos de id con .replace pero se mantiene su tipo como str
dfUnido["ID_aux"] = dfUnido["ID_aux"].replace("id_", "", regex=True)
dfUnido
```

```python
# se reemplazan los posibles valores "vacios" con NaN pero no se eliminan debido a la gran cantidad de valores que pueden haber
# y no nos interesa quedarnos sin datos, solo saber cuales son NaN
dfUnido = dfUnido.replace('*', np.nan)
dfUnido = dfUnido.replace('None', np.nan)
dfUnido = dfUnido.replace(' ', np.nan)
dfUnido = dfUnido.replace(-999, np.nan)
```
