```python
dfUnido = pd.concat([df2023_filtrado, df2022_filtrado, df2021_filtrado, df2020_filtrado, df2019_filtrado], axis=0, join="outer").reset_index(drop=True)
dfUnido
```

```python
# Reemplazamos los posibles valores "vacios con NaN pero no los eliminaremos, ya que se eliminaria gran cantidad de datos 
# importantes y  nuestro interes es saber cuales son nulos
dfUnido = dfUnido.replace('*', np.nan)
dfUnido = dfUnido.replace('None', np.nan)
dfUnido = dfUnido.replace(' ', np.nan)
dfUnido = dfUnido.replace(-999, np.nan)
```

# Filtraciones y Cambios Generales

```python
# se cambia la primera parte de los codigos de id con .replace pero se mantiene su tipo como str
dfUnido["ID_aux"] = dfUnido["ID_aux"].replace("id_", "", regex=True)
```

```python
# Cambiamos el tipo de datos
dfUnido['SITUACION_POSTULANTE'] = dfUnido['SITUACION_POSTULANTE'].astype('category')
dfUnido
```

```python
dfUnido.to_csv("Dataframe_postulaciones.csv", index=False)
```
